```python
#importing basic libraries-
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
df=sns.load_dataset("iris") #data ingestion
```

```python
#removing duplicate value-
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
#Mean value of sepal_length,sepal_width,petal_length,petal_width for each different species-
df.groupby(["species"])[["sepal_length","sepal_width","petal_length",
                        "petal_width"]].mean()
```

```python
#count of sepal_length,sepal_width,petal_length,petal_width for each different species-
df.groupby(["species"])[["sepal_length","sepal_width","petal_length",
                        "petal_width"]].count()
```

# Bivariate analysis & Multivariate analysis--¶

```python
sns.scatterplot(data=df,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.catplot(x="species",y="sepal_width",kind="swarm",data=df)
```

```python
sns.distplot(df[df["species"]=="setosa"]["sepal_length"],hist=False,color="red")
sns.distplot(df[df["species"]=="versicolor"]["sepal_length"],hist=False,color="blue")
sns.distplot(df[df["species"]=="virginica"]["sepal_length"],hist=False,color="green")
```
